### Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.metrics import mean_squared_error as mse
```

### Loading the data

```python
df = pd.read_csv('final_data_.csv')
```

### seeing the top 5 rows of data using head

```python
df.head()
```

### Deleting unwanted column using drop method

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

### seeing last 5 rows of data using tail 

```python
df.tail()
```

### seeing shape of the data to check how many rows and how many columns are present in the data using shape

```python
df.shape
```

### Checking the null values

```python
df.isna().sum()
```

### Checking the datatypes of the columns

```python
df.dtypes
```

### Converting the datatypes

```python
df['ingredient_type'] = df['ingredient_type'].astype('category') 
df['date'] = pd.to_datetime(df['date'])
```

### Again checking the datatypes of the columns

```python
df.dtypes
```

### Setting the date to index

```python
df1 = df.set_index("date")
```

### Dropping the unwanted columns

```python
df2 = df1.drop(['farm_area','farming_company','deidentified_location'],axis=1)
```

### Creating new features using index

```python
def create_features(df2):
    df2.copy()
    df2['hour'] = df2.index.hour
    df2['dayofweek'] = df2.index.dayofweek
    df2['quarter'] = df2.index.quarter
    df2['month'] = df2.index.month
    df2['dayofyear'] = df2.index.dayofyear
    return df2
```

```python
df2 = create_features(df2)
```

### Checking the top 5 rows of data

```python
df2.head()
```

### Replacing the ingredient_types with values

```python
df2['ingredient_type']=df2['ingredient_type'].replace({'ing_w':0,'ing_x':1,'ing_y':2,'ing_z':3})
```

### Plotting the 

```python
fig = plt.figure(figsize = (10, 5))
plt.plot(df2['month'], df2['yield'])
plt.xlabel("month")  
plt.ylabel("yield")  
plt.title("Yield based on month")  
plt.show()
```

### Storing the independent variables in X and  the dependent  in y & dropping the id column

```python
X = df2.drop(['farm_id','yield'],axis=1)
```

```python
y= df2['yield']
```

### Splitting the data  using train_test_split

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
print('Y Target Size:', len(y_train))
print('X Train Size:', len(X_train))
print('X valid  Size:', len(X_valid))
print('Y Target Size:', len(y_valid))
```

### storing the numerical columns in num_cols

```python
num_cols= ['temp_obs','wind_direction','dew_temp', 'pressure_sea_level', 'precipitation','wind_speed']
```

```python
X_train.head()
```

### Using Standard scaling to scale the numerical features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X[num_cols] = scaler.fit(X[num_cols])
X_train[num_cols] = scaler.transform(X_train[num_cols])
X_valid[num_cols] = scaler.transform(X_valid[num_cols])
```

```python
X_train.head()
```

### Building the RandomForestRegressor model

```python
reg_rf = RandomForestRegressor()
reg_rf.fit(X_train,y_train)
```

### Predicting the train & test data using trained model

```python
y_train_pred = reg_rf.predict(X_train) 
```

```python
y_valid_pred = reg_rf.predict(X_valid)
```

### Evaluating the model using metrics

```python
np.sqrt(metrics.mean_squared_error(y_train,y_train_pred))
```

```python
np.sqrt(metrics.mean_squared_error(y_valid,y_valid_pred))
```

### Using Xgboost regressor to fit the train data

```python
xgb = XGBRegressor()
res = xgb.fit(X_train,y_train)
```

```python
res
```

### Predicting the train & test data using trained model

```python
y_train_pred_1 = xgb.predict(X_train) 
```

```python
y_valid_pred_1 = xgb.predict(X_valid)
```

### Evaluating the model using metrics

```python
np.sqrt(metrics.mean_squared_error(y_train,y_train_pred_1))
```

```python
np.sqrt(metrics.mean_squared_error(y_valid,y_valid_pred_1))
```

### Knowing the importance of each feature 

```python
importance = xgb.get_booster().get_score(importance_type='weight')
```

```python
for feature, score in importance.items():
    print('{}: {}'.format(feature,score))
```
